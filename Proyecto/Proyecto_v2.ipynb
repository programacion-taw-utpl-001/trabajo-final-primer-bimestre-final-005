```python
import pandas as pd
```

```python
data = pd.read_csv('camas_2016_3-2.csv',engine = 'python')
data
```

```python
#Presentamos los nombres de las columnas para eliminar las que consideremos innecesarias
data.columns
```

```python
data.pop('PARROQUIA_UBICACION')
data.pop('CLASE_ESTABLECIMIENTO')
data.pop('TIPO_ESTABLECIMIENTO')
data.pop('ENTIDAD')
data.pop('SECTOR')
data.pop('MEDICINA')
data.pop('CIRUGIA')
data.pop('GINECOLOGIA_OBSTETRICIA')
data.pop('PEDIATRIA')
data.pop('CARDIOLOGIA')
data.pop('NEUMOLOGIA')
data.pop('PSIQUIATRIA')
data.pop('TRAUMATOLOGIA')
data.pop('INFECTOLOGIA')
data.pop('OFTALMOLOGIA_OTORRINO')
data.pop('UROLOGIA')
data.pop('GASTROENTEROLOGIA')
data.pop('OTROS_SERVICIOS')
data.pop('SERVICIOS_INDIFERENCIADOS')
data.pop('CAMAS_EMERGENCIA')
data.pop('CAMAS_CUIDADOS_INTENSIVOS')
data.pop('CAMAS_DISPONIBLES_EMERGENCIA')
data.pop('CAMAS_DISPONIBLES_CUIDADOS_INTENSIVOS')
data.pop('TOTAL_CAMAS_DONACION_NORMAL')
data
```

```python
#Obtenemos la cantidad de camas disponibles al nivel del pais
Total_camas_disponibles_pais = (data['TOTAL_CAMAS_DISPONIBLES'].astype(int)).sum()
print(Total_camas_disponibles_pais)
```

```python
dataL = data[data['PROVINCIA_ESTABLECIMIENTO'] == 'Loja']
dataLoja = dataL[dataL['CANTO_ESTABLECIMIENTO'] == 'Loja']
dataLoja
```

```python
#Obtenemos la cantidad de camas disponibles al nivel de la Provincia Loja
Total_camas_disponibles_prov_loja = (dataL['TOTAL_CAMAS_DISPONIBLES'].astype(int)).sum()
print(Total_camas_disponibles_prov_loja)
```

```python
#Obtenemos la cantidad de camas disponibles al nivel de la cantÃ³n Loja
Total_camas_disponibles_loja = (dataLoja['TOTAL_CAMAS_DISPONIBLES'].astype(int)).sum()
print(Total_camas_disponibles_loja)
```

```python
#Porcentaje de camas que posee la Provincia de Loja
porcentaje_camas_prov_loja = ((Total_camas_disponibles_prov_loja*100)/Total_camas_disponibles_pais)
print (porcentaje_camas_prov_loja)
```

```python
#Porcentaje de camas que posee la cantÃ³n de Loja
porcentaje_camas_loja = ((Total_camas_disponibles_loja*100)/Total_camas_disponibles_pais)
print (porcentaje_camas_loja)
```

```python
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import RdYlGn3
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

output_file("camas_totales.html")

camas = ['PaÃ­s', 'Provincia Loja', 'CantÃ³n Loja']
cantidad = [100, porcentaje_camas_prov_loja, porcentaje_camas_loja]

source = ColumnDataSource(data=dict(camas=camas, cantidad=cantidad))

p = figure(x_range=camas, plot_height=500, toolbar_location=None, title="Total Camas")
p.vbar(x='camas', top='cantidad', width=0.9, source=source, legend="camas",
       line_color='white', fill_color=factor_cmap('camas', palette=RdYlGn3, factors=camas))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 100
p.legend.orientation = "horizontal"
p.legend.location = "top_right"

show(p)
```

```python
#Obtenemos la cantidad de dias disponibles al nivel del cantÃ³n Loja
Total_dias_disponibles_loja = (dataLoja['DIAS_CAMAS_DISPONIBLES'].astype(int)).sum()
print(Total_dias_disponibles_loja)
```

```python
#Obtenemos la cantidad de dias ocupados al nivel del cantÃ³n Loja
Total_dias_ocupados_loja = (dataLoja['DIAS_DE_ESTADA'].astype(int)).sum()
print(Total_dias_ocupados_loja)
```

```python
from bokeh.palettes import Accent3
output_file("dias_ocupados_loja.html")

porcentaje_dias_ocupados_loja=((Total_dias_ocupados_loja*100)/Total_dias_disponibles_loja)

dias = ['DÃ­as Disponibles', 'DÃ­as Ocupados']
cantidad = [100, porcentaje_dias_ocupados_loja]

source = ColumnDataSource(data=dict(dias=dias, cantidad=cantidad))

p = figure(x_range=dias, plot_height=500, toolbar_location=None, title="Dias Ocupados")
p.vbar(x='dias', top='cantidad', width=0.9, source=source, legend="dias",
       line_color='white', fill_color=factor_cmap('dias', palette=Accent3, factors=dias))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 100
p.legend.orientation = "horizontal"
p.legend.location = "top_right"

show(p)
```

```python
#Datos de las camas individuales de Loja
camas_medicina = (dataLoja['CAMAS_DISPONIBLES_MEDICINA'].astype(int)).sum()
camas_cirugia = (dataLoja['CAMAS_DISPONIBLES_CIRUGIA'].astype(int)).sum()
camas_ginecologia = (dataLoja['CAMAS_DISPONIBLES_GINECOLOGIA'].astype(int)).sum()
camas_pediatria = (dataLoja['CAMAS_DISPONIBLE_PEDIATRIA'].astype(int)).sum()
camas_cardiologia = (dataLoja['CAMAS_DISPONIBLES_CARDIOLOGIA'].astype(int)).sum()
camas_neumologia = (dataLoja['CAMAS_DISPONIBLES_NEUMOLOGIA'].astype(int)).sum()
camas_psiquiatria = (dataLoja['CAMAS_DISPONIBLE_PSIQUIATRIA'].astype(int)).sum()
camas_traumologia = (dataLoja['CAMAS_DISPONIBLE_TRAUMOLOGIA'].astype(int)).sum()
camas_infectologia = (dataLoja['CAMAS_DISPONIBLES_INFECTOLOGIA'].astype(int)).sum()
camas_oftalmologia = (dataLoja['CAMAS_DISPONIBLES_OFTALMOLOGIA'].astype(int)).sum()
camas_urologia = (dataLoja['CAMAS_DISPONIBLES_UROLOGIA'].astype(int)).sum()
camas_gastroenterologia = (dataLoja['CAMAS_DISPONIBLES_GASTROENTEROLOGIA'].astype(int)).sum()
camas_otros_servicios = (dataLoja['CAMAS_DISPONIBLES_OTROS_SERVICIOS'].astype(int)).sum()
camas_servicios_ind = (dataLoja['CAMAS_DISPONIBLES_SERVICIOS_INDIFERENCIADOS'].astype(int)).sum()

camas_total_indi_loja = (camas_medicina + camas_cirugia +camas_ginecologia +camas_pediatria+  camas_cardiologia+  camas_neumologia+ camas_psiquiatria+ camas_traumologia+ camas_infectologia+ camas_oftalmologia+ camas_urologia+ camas_gastroenterologia+ camas_otros_servicios+ camas_servicios_ind)
print (camas_total_indi_loja)
```

```python
from bokeh.palettes import inferno
output_file("camas_individuales.html")


camas = ['Medicina', 'Cirugia','Ginecologia','Pediatria','Cardiologia','Neumologia','Psiquiatria','Traumatologia','Infectologia','Oftalmologia','Urologia','Gastroenterologia','Otros Servicios','Servicios Indiferenciados']
cantidad = [camas_medicina, camas_cirugia, camas_ginecologia, camas_pediatria, camas_cardiologia, camas_neumologia, camas_psiquiatria, camas_traumologia, camas_infectologia, camas_oftalmologia, camas_urologia, camas_gastroenterologia, camas_otros_servicios, camas_servicios_ind]

source = ColumnDataSource(data=dict(camas=camas, cantidad=cantidad))

p = figure(x_range=camas, plot_height=600, plot_width = 1600, toolbar_location=None, title="Cantidad de Camas Individuales")
p.vbar(x='camas', top='cantidad', width=0.2, source=source, legend="camas",
       line_color='white', fill_color=factor_cmap('camas', palette=inferno(15), factors=camas))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 200
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

```python
fallecidos_menor_48h = (dataLoja['NUMERO_FALLECIDOS_MENOR_48H'].astype(int)).sum()
print(fallecidos_menor_48h)
```

```python
fallecidos_mayor_48h = (dataLoja['NUMERO_FALLECIDOS_EN_48H_Y_MAS'].astype(int)).sum()
print(fallecidos_mayor_48h)
```

```python
from matplotlib import pyplot

output_file('fallecidos.html')
pyplot.savefig('fallecidos.png')
fallecidos=('Menor a 48h','en 48h y mas')
slices=(fallecidos_menor_48h,fallecidos_mayor_48h)
colores=('red','blue')
pyplot.pie(slices,shadow=True,labels=fallecidos, colors=colores,autopct=('%1.1f%%'))
pyplot.title('Fallecidos',bbox={"facecolor":"1.8", "pad":5})

pyplot.show()
```
