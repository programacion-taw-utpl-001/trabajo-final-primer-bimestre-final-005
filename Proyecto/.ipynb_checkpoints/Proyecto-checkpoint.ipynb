```python
data = pd.read_csv('camas_2016_2.csv',engine = 'python')
data
```

```python
data.columns
```

```python
data.pop('PARROQUIA_UBICACION')
data.pop('CLASE_ESTABLECIMIENTO')
data.pop('TIPO_ESTABLECIMIENTO')
data.pop('ENTIDAD')
data.pop('SECTOR')
data.pop('CAMAS_DISPONIBLES_EMERGENCIA')
data.pop('CAMAS_DISPONIBLES_CUIDADOS_INTENSIVOS')
data
```

```python
dataL = data[data['PROVINCIA_ESTABLECIMIENTO'] == 'Loja']
dataLL = dataL[dataL['CANTO_ESTABLECIMIENTO'] == 'Loja']
dataLL
```

```python
data_Lo = dataLL.set_index('DIAS_CAMAS_DISPONIBLES')
data_Lo
```

```python
Total_dias_disponibles_loja = (dataLL['DIAS_CAMAS_DISPONIBLES'].astype(float)).sum()
print(Total_dias_disponibles_loja)
```

```python
data_Lo = dataLL.set_index('DIAS_DE_ESTADA')
data_Lo
```

```python
Total_dias_estadia_loja = (dataLL['DIAS_DE_ESTADA'].astype(float)).sum()
print(Total_dias_estadia_loja)
```

```python
#Porcentaje de dias ocupados en cama vs dias totales disponibles
porcentaje_dias = ((Total_dias_estadia_loja*100)/Total_dias_disponibles_loja)
from bokeh.plotting import figure, output_file, show

plot = figure(plot_width=1000, plot_height=300)
plot.hbar(y=[0.1, 0.5], height=0.1, left=0, right=[100,porcentaje_dias], color= "red" )

show(plot)
```
